```python
digits = load_digits()
img_path = './img/'
bground_list = os.listdir(img_path)
img_data =[]
```

```python
chars = pd.DataFrame([])
faces = []
for i in bground_list:
    img = Image.open(img_path+i)
    img = img.resize((64, 64))
    img = np.array(img)
    img = img / 255
    char = pd.Series(img.flatten(),name=i)
    chars = chars.append(char)
```

```python
from sklearn.decomposition import PCA

chars_pca = PCA(n_components=300) # 몇개의 주성분을 쓸 것인지(글자 갯수만큼 있어야 되나?)
chars_pca.fit(chars)
# fig, axes = plt.subplots(2,10,figsize=(9,3),
# subplot_kw={'xticks':[], 'yticks':[]},
# gridspec_kw=dict(hspace=0.01, wspace=0.01))

# for i, ax in enumerate(axes.flat):
#      ax.imshow(chars_pca.components_[i].reshape(64,64),cmap='gray')
```

```python
components = chars_pca.transform(chars) # 특징행렬을 낮은 차원의 근사행렬로 변환

plt.scatter(components[:, 0], components[:, 1], c= '0',s=10, cmap='viridis')
projected = chars_pca.inverse_transform(components) #변환된 근사행렬을 원래의 차원으로 복귀
# fig, axes = plt.subplots(10,10,figsize=(9,9), subplot_kw={'xticks':[], 'yticks':[]},
#             gridspec_kw=dict(hspace=0.01, wspace=0.01))

# for i, ax in enumerate(axes.flat):
#     ax.imshow(projected[i].reshape(64,64),cmap="gray")

```

```python
def viz_img(y_pred):
    n = 10
    fig = plt.figure(1)
    box_index = 1
    for cluster in range(190):
        result = np.where(y_pred == cluster)
        for i in np.random.choice(result[0].tolist(), n, replace=True):
            ax = fig.add_subplot(n, n, box_index)
            plt.imshow(projected[i].reshape(64, 64))
            plt.gray()
            ax.get_xaxis().set_visible(False)
            ax.get_yaxis().set_visible(False)
            box_index += 1
    plt.show()
```

```python
def filename(frame_id):
        return "%08d.jpg" % frame_id
```

```python
def save_img(y_pred):
    label_ids = np.unique(y_pred) # 라벨이 저장
    num_unique_faces = len(np.where(label_ids > -1)[0]) # 라벨의 갯수
    for label_id in label_ids:
        dir_name = "ID%d" % label_id
        if not os.path.isdir(dir_name):
                os.mkdir(dir_name)     
        indexes = np.where(y_pred == label_id)[0]
        for i in indexes:
            pathname = os.path.join('./img/', filename(indexes))
            image = cv2.imread(pathname)
            filename = dir_name + "-" + filename(indexes)
            pathname = os.path.join(dir_name, filename)
            cv2.imwrite(pathname, image)
            
            
```

```python
def plot_kmeans(kmeans, X, n_clusters=190, rseed=0, ax=None):
    labels = kmeans.fit_predict(X)

    # plot the input data
    ax = ax or plt.gca()
    ax.axis('equal')
    ax.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis', zorder=2)

    # plot the representation of the KMeans model
    centers = kmeans.cluster_centers_
    radii = [cdist(X[labels == i], [center]).max()
             for i, center in enumerate(centers)]
    for c, r in zip(centers, radii):
        ax.add_patch(plt.Circle(c, r, fc='#CCCCCC', lw=3, alpha=0.5, zorder=1))
```
